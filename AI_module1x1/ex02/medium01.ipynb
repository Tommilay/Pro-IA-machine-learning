```python
data = df.drop(columns='index')
data
```

```python
x = data.drop(columns='type')
y = data.type

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=72)
```

```python
print('Class istribution im training data')
print(y_train.value_counts())
print('Class distribution in test data')
print(y_test.value_counts())
```

```python
plt.pie(y_train.value_counts(), autopct="%1.1f%%")
plt.show()
```

```python
from numpy.random import multinomial
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
log_reg = LogisticRegression(max_iter=1000, multi_class='multinomial')
log_reg.fit(x_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

train_predictions = log_reg.predict(x_train_scaled)

train_accuracy = accuracy_score(y_train, train_predictions)
train_cm = confusion_matrix(y_train, train_predictions)

print(f"Accuracy om training data : {train_accuracy}")
print(train_cm)
```

```python
test_predictions = log_reg.predict(x_test_scaled)

test_accuracy = accuracy_score(y_test, test_predictions)
test_cm = confusion_matrix(y_test, test_predictions)

print(f"Accuracy on test data : {test_accuracy}")
print(test_cm)
```

```python
sns.heatmap(test_cm, annot=True, cmap='Blues')
```
