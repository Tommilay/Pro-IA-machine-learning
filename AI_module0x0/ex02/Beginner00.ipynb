```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Salary_dataset.csv")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X=data[["YearsExperience"]]
Y=data["Salary"]

regression = LinearRegression()

regression.fit(X,Y)

predictions = regression.predict([[10]])

print("Predicted salary for 10 years of experience : ", predictions[0])
```
